## Las variables
1. X --> Cantidad de lapices
1. PAG --> Pago que se realizara por lapices

## Formulas
1. $X > = 1000$
2. $PAG = X*0.85$
3. $PAG = X*0.90$

## Diagrama de flujo con Lucichart
<img src="https://raw.githubusercontent.com/MarioMarentes128/Act-12-p3.3-Pago-por-equis-cantidad-de-lapices/refs/heads/main/DF%20usando%20if%20else.png" width="400" height="400" alt="Descripción de la imagen">

```python
# para cuando es verdadero (si)
# calculando la cantidad de lapices usando if else
# Datos de entrada X
# mostrar el pago segun la decicion

X = int(input("Ingrese la cantidad de lapices: "))
if X < 1000:
    PAG = X * 0.90
else:
    PAG = X * 0.85
    print("El pago por lapices es: ", PAG)
print("Fin del programa")
print("Programa realizado por Mario Yañez NL = 53")
```

```python
# para cuando es falso (No)
# calculando la cantidad de lapices usando if else
# Datos de entrada X
# mostrar el pago segun la decicion
X = int(input("Ingrese la cantidad de lapices: "))
if X < 1000:
    PAG = X * 0.90
else:
    PAG = X * 0.85
print("El pago por lapices es: ", PAG)
print("Fin del programa")
print("Programa realizado por Mario Yañez NL = 53")
```
